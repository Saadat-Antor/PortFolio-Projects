```python
#IMPORTING NECESSARY LIBRARIES

import pandas as pd
import seaborn as sbn
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure as fig

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusting the figure configuration


#READING THE DATASET

df = pd.read_csv(r'C:\Users\Sadat\Desktop\Portfolio\Projects\Data Analyst Portfolio Project- Alex The Analyst\04. DATA CORRELATION IN PYTHON\movies.csv')
```

```python
#LOOKING INTO THE NEWLY CREATED DATAFRAME

df.head()

```

```python
# DETERMINING MISSING DATA 

#df.info()

df.dropna(inplace = True) #dropping the rows that has missing dataa
for col in df.columns:
    perc_missing_col = np.mean(df[col].isnull())
    print('{} - {}%' .format(col, perc_missing_col))
```

```python
# LOOKING INTO DATA TYPES (COLUMN)

df.dtypes
```

```python
# MODIFYING DATA TYPES OF COLUMN (BUDGET, GROSS)

df.budget = df.budget.astype('int64')
df.gross = df.gross.astype('int64')
df.votes = df.votes.astype('int64')
#df
```

```python
# LOOKING AT ALL THE DATA IN THE DATAFRAME

pd.set_option('display.max_rows',None)
```

```python
# SORTING THE DATAFRAME 

df_02 = df.sort_values(by=['gross'],inplace=False, ascending=False)
df_02.head()
```

```python
# DROPPING ALL THE DUPLICATE DATA

df_03 = df_02.drop_duplicates()
df_03.head()
```

```python
# HYPOTHESIS 1 = THE HIGHER THE BUDGET, HIGHER THE GROSS INCOME


# Creating a scattered plot for initial understanding

plt.scatter(x = df_03.budget, y = df_03.gross, color = 'lime')

plt.title('GROSS VS BUDGET')
plt.xlabel('MOVIE BUDGET (IN HUNDRED MILLIONS)')
plt.ylabel('MOVIE GROSS INCOME (IN HUNDRED MILLIONS)')

plt.show()
```

```python
# REGRESSION PLOT USING SEABORN

sbn.regplot(x = df_03.budget, y = df_03.gross, data = df_03, scatter_kws = {'color': 'lime'}, line_kws = {'color': 'darkgreen'})
```

```python
# EXPLORING DIFFERENT CORRELATION METHODS

#pearson method
df_corr = df_03.corr(method = 'pearson')
df_corr
```

```python
#kendall method
df_corr_02 = df_03.corr(method = 'kendall')
df_corr_02
```

```python
#spearman method
df_corr_03 = df_03.corr(method = 'spearman')
df_corr_03
```

```python
#TESTING HYPOTHESIS 01 USING DIFFERENT METHODS OF CORRELATION

sbn.heatmap(df_corr, annot=True)

plt.title('CORRELATION BETWEEN MOVIE FEATURES (PEARSON METHOD)')
plt.xlabel('MOVIE FACTORS')
plt.ylabel('MOVIE FACTORS')

plt.show()
```

```python
sbn.heatmap(df_corr_02, annot=True)

plt.title('CORRELATION BETWEEN MOVIE FEATURES (KENDALL METHOD)')
plt.xlabel('MOVIE FACTORS')
plt.ylabel('MOVIE FACTORS')

plt.show()
```

```python
sbn.heatmap(df_corr_03, annot=True)

plt.title('CORRELATION BETWEEN MOVIE FEATURES (SPEARMAN METHOD)')
plt.xlabel('MOVIE FACTORS')
plt.ylabel('MOVIE FACTORS')

plt.show()
```

```python
df_03.head()
```

```python
# CATEGORIZING THE COLUMNS WITH OBJECT DATATYPE NUMERICALLY

df_catg_num = df_03
for col in df_catg_num.columns:
    if (df_catg_num[col].dtype == 'object'):
        df_catg_num[col] = df_catg_num[col].astype('category')
        df_catg_num[col] = df_catg_num[col].cat.codes
    
df_catg_num.head()
```

```python
# CORRELATION BETWEEN ALL MOVIE FACTORS
df_catg_num_corr = df_catg_num.corr(method = 'pearson')
df_catg_num_corr
```

```python
###CORRELATION BETWEEN ALL MOVIE FACTORS WITH THE HEATMAP

df_catg_num_corr = df_catg_num.corr(method = 'pearson')

sbn.heatmap(df_catg_num_corr, annot=True)

plt.title('CORRELATION BETWEEN MOVIE FEATURES (PEARSON METHOD)')
plt.xlabel('MOVIE FACTORS')
plt.ylabel('MOVIE FACTORS')

plt.show()
```

```python
# FOR BETTER UNDERSTANDING THE CORRELATION

df_catg_num_corr.unstack()
```

```python
# SORTING THE DATA

df_catg_num_corr_sorted = df_catg_num_corr.sort_values(by=['name'])
print (df_catg_num_corr_sorted)
```

```python
# CORRELATION MORE THAN 0.3

print(df_catg_num_corr_sorted[abs(df_catg_num_corr_sorted) > 0.5])

```
