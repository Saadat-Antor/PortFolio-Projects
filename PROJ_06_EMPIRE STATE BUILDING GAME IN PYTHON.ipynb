# Imagine you're walking up the stairs of the empire state building and you're playing a game with your best friend. You throw a dice one hundred times. The conditions are:

# •	If it's 1 or 2 you'll go one step down.
# •	If it's 3, 4, or 5, you'll go one step up.
# •	If you throw a 6, you have to throw the dice again and walk up the resulting number of steps.

# QUESTION: What are the odds that you'll reach 60 steps high on the Empire State Building?

# NOTE: YOU'RE A BIT CLUMSY AND YOU HAVE A 0.2% CHANCE OF FALLING DOWN.


## RANDOM FLOAT NUMBERS FOR ROLLING A DICE

```python
# Importing numpy as np
import numpy as np

# Setting the seed
np.random.seed(123)

# Generating and printing random a float number
print(np.random.rand())
```

## ROLLING A DICE

```python
# Using randint() to simulate a dice
np.random.randint(1, 7)

# Using randint() again
print(np.random.randint(1, 7))
print(np.random.randint(1, 7))
print(np.random.randint(1, 7))
print(np.random.randint(1, 7))
print(np.random.randint(1, 7))
print(np.random.randint(1, 7))
```

## SIMULATING NUMBER OF STEPS AFTER ROLLING A DICE

```python
# Starting step
step = 50 

# Rolling the dice
dice = np.random.randint(1,7)

# Finishing the control construct
if dice <= 2 :
    step = step - 1
elif dice >=3 and dice <=5 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

# Printing out dice and step
print(dice)
print(step)
```

## SIMUALTING RANDOM WALK (COLLECTION OF STEPS)

```python
# Initializing random_walk
random_walk = [0]

for x in range(100) :
    # Setting step
    step = random_walk[-1]

    # Rolling the dice
    dice = np.random.randint(1,7)

    # Determining the next step
    if dice <= 2:
        step = max(0,step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # appending the next_step to random_walk
    random_walk.append(step)

# Printing random_walk
print(random_walk)
```

## VISUALIZING THE WALK

```python
# Importing matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plotting random_walk
plt.plot(random_walk)
plt.title('RANDOM WALK')

# Showing the plot
plt.show()
```

## SIMULATING MULTIPLE WALKS

```python
# Simulating random walk 500 times
all_walks = []

for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implementing clumsiness because you're a bit clumsy and you have a 0.2% chance of falling down.
        if np.random.rand() <= 0.002 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)
print(all_walks)
```

## Visualize all walks

```python
# Creating and plotting np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

## PLOTTING A DISTRIBUTION FOR ALL_WALKS

```python
# Selecting last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plotting histogram of ends, display plot
plt.hist(ends)
plt.show()
```
